# Exercise Analysis Pipeline Documentation


### 1. Introduction

This project aims to develop a **Computer Vision-based exercise analysis system** that evaluates a user's exercise performance using **MediaPipe BlazePose** for human pose estimation.

The system extracts body landmarks from exercise videos, computes joint angles and motion features, compares the performed movement against exercise-specific rules stored in JSON configuration files, and finally provides repetition counting and performance feedback.

---

### 2. Project Objectives

The main objectives of this project are:

* Detect human body pose using MediaPipe BlazePose.
* Calculate major joint angles.
* Extract motion-related features.
* Count exercise repetitions.
* Evaluate exercise performance.
* Generate user feedback.
* Support multiple exercises through independent JSON configuration files.

---

### 3. System Overview

The system is composed of several independent modules, where each module is responsible for a specific task.

```text
Select Exercise
        │
        ▼
Load Exercise Configuration (JSON)
        │
        ▼
Load Video
        │
        ▼
Pose Detector
        │
        ▼
Angle Calculator
        │
        ▼
Feature Extraction
        │
 ┌──────┴──────┐
 ▼             ▼
Repetition   Feedback
 Counter      Engine
        │
        ▼
Display Results
```

### 5. Module Description

#### 5.1 Dataset Loader

Responsible for:

* Reading all available exercises.
* Loading reference videos.
* Loading user videos.
* Returning the required file paths.

---

#### 5.2 Pose Detector

This module is built on **MediaPipe BlazePose**.

Responsibilities:

* Read video frames.
* Detect the human body.
* Extract 33 body landmarks.
* Extract 3D world landmarks.

---

#### 5.3 Angle Calculator

Calculates the angles of major body joints, including:

* Elbow
* Shoulder
* Hip
* Knee

The module uses **3D landmark coordinates** to calculate joint angles.

---

#### 5.4 Feature Extraction

Extracts motion characteristics from the calculated angles, including:

* Range of Motion (ROM)
* Minimum Angle
* Maximum Angle
* Average Angle
* Movement Direction
* Angular Speed
* Stability
* Exercise Duration

---

#### 5.5 Repetition Counter

Uses the rules defined in the exercise JSON configuration file.

Responsibilities:

* Track the selected joint angle.
* Detect movement start.
* Detect movement completion.
* Count exercise repetitions.

---

#### 5.6 Feedback Engine

Compares the calculated joint angles against the predefined exercise rules stored in the JSON configuration file.

Outputs include:

* Performance Score
* Detected Errors
* Feedback Messages

---

### 6. Exercise Dataset Structure

Each exercise folder contains:

```text
Exercise Folder

│

├── reference/

├── user/

└── exercise_config.json
```

---

### 7. Exercise Configuration File (exercise_config.json)

Each configuration file contains all exercise-specific information, including:

* Exercise name
* Target muscles
* Joint angle ranges
* Repetition counting rules
* Common execution errors
* Feedback messages

This design allows new exercises to be added without modifying the source code. Only a new configuration file and corresponding videos are required.

---

### 8. System Workflow

The execution pipeline follows these steps:

1. The user selects an exercise.
2. The user selects the video type (Reference or User).
3. The desired video is selected.
4. The corresponding JSON configuration file is loaded.
5. Pose detection is performed.
6. Joint angles are calculated.
7. Motion features are extracted.
8. Exercise repetitions are counted.
9. Performance is evaluated.
10. Final results are displayed.

---

### 9. Current Output

The system currently reports:

* Total number of video frames
* Number of successfully detected frames
* Video frame rate (FPS)
* Number of repetitions
* Best frame selected for evaluation
* Performance score
* Detected execution errors
* Feedback messages
* Joint feature analysis table
